## Data Quality Issues in the Dataset



1. **Null or Missing Values**

   - These occur when some entries in the dataset are empty or not recorded. Missing values can affect analysis and model performance. It's important to check for nulls and decide whether to fill, drop, or otherwise handle them.



2. **Duplicate Rows**

   - Duplicate records arise from repeated data entry or merging datasets. Duplicates can bias results and should be identified and removed if not intentional.



3. **Inconsistent Data Types**

   - this occurs when columns may have mixed types (e.g., numbers stored as strings, or dates as text). This can cause errors in analysis and should be standardized (e.g., converting all dates to datetime objects).
